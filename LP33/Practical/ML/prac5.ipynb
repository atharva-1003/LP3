```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=pd.read_csv("diabetes.csv")
data
```

```python
data.dtypes
```

```python
x=data.drop(columns=["Outcome"],axis=1)
x
```

```python
y=data["Outcome"]
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
'''

What is StandardScaler?
The StandardScaler is a feature scaling technique used in machine learning to standardize or normalize the features 
(input data) by removing the mean and scaling the data to unit variance. It is commonly used when the features have 
different units (e.g., height in meters, weight in kilograms) or when the data is not centered or scaled. Standardization 
ensures that all features contribute equally to the model's performance and helps certain machine learning algorithms perform 
better.
'''
```

```python
standardizer=StandardScaler()
xtrain_processed=standardizer.fit_transform(xtrain)
xtest_processed=standardizer.fit_transform(xtest)
xtrain_processed
```

```python
knn_model=KNeighborsClassifier(3)
knn_model.fit(xtrain_processed,ytrain)
```

```python
ypred=knn_model.predict(xtest_processed)
ypred
```

```python
cm=confusion_matrix(ytest,ypred)
accuracy=accuracy_score(ytest,ypred)
precision=precision_score(ytest,ypred)
f1=f1_score(ytest,ypred)
recall=recall_score(ytest,ypred)
print(cm)
accuracy,precision,f1,recall
```
